## Start with some Feature Enginering to Timestamp

```python
# convert  the column from strings to DateTime (before this we try use x.strftine but it didn't worked)
train['timestamp']=pd.to_datetime(train['timestamp']) 
type(train['timestamp'])
```

```python
train['timestamp']=pd.to_datetime(train['timestamp']) # convert all data in this col to datetime
train['Year'] = train['timestamp'].map(lambda x: x.year) # get the year 
train['Month']=train['timestamp'].map(lambda x: x.month)# get  month 
train['Day of Week']=train['timestamp'].map(lambda x: x.day) # the  day in a manth 
```
