# Sberbank Russian Housing Market

**Can you predict realty price fluctuations in Russia’s volatile economy?**

### Introduction 

Buying a house is an expensive task, whether it's for living or investing. 
Those who want to buy a house would want to make sure they pay the right price for the house.

There are many factors that might influence the selling price of a property: House properties, like area, number of rooms and the age of the building; Neighbourhood properties, like distance from city center or from schools, and district properties, like distance from the capital.

The current state of the economy also influence the price. Unstable economy could lead a property price to vary over time, which make it more difficult to determine whether the price is reasonable.

Therefore, having a way to evaluate the price of a property from its features, considering the current economic state, could help many people be more confident about making the decision of buying the property.  

The Sberbank housing market data is a rich dataset, that includes housing data and macroeconomic patterns. Sberbank is Russia’s oldest and largest bank, and the dataset they published contains many transction records. 

Using the Sberbank dataset we want to try and build a good price prediction model for the russian properties.

The Sberbank dataset was taken from the Kaggle competition "Sberbank Russian Housing Market". https://www.kaggle.com/c/sberbank-russian-housing-market

### Project goal

We want to crate a model for predicting the price of a given property. The model will evaluate a target price for the property, regarding the property features, that include the house properties and information about the surronding neighborhood. The model will also consider the macroeconomic information of the date of the transaction.

## Part 3- Price Prediction model 

### Model A- using scikit learn

First, we tried to build a prediction model using Scikit-learn gradient boosting package.

** Import required packages **

```python
from sklearn import ensemble
from sklearn import datasets
from sklearn.utils import shuffle
from sklearn.metrics import mean_squared_error
```

** Seperate the price (the feature the model should predict) from the rest of the data set **

```python
Y=expend_train['price_doc']/100000 #normalizing the destination make better results
X=expend_train.drop(['id','full_timestamp','price_doc'],axis=1)
```

** Drop the object features - only numerical would go into the gradient boosting**

```python
columnames=np.array(X.columns.values)
objfilter=np.array(list(X.dtypes==object),dtype=bool)
objnames=columnames[objfilter]
X = X.drop(list(objnames),axis=1)
```

** Impute NaN values ** 

```python
from sklearn.preprocessing import Imputer
# Creat out impute to replace missing values with the mean
imp=Imputer(missing_values='NaN', strategy='median', axis=0)

# Impute our data
imp_X=pd.DataFrame(imp.fit_transform(X))
imp_X.columns=X.columns
imp_X.index=X.index
```

```python
#imp_X.head().transpose()
```

** Arrange the data for prediction **

```python
# devide the data to train and test randomly
import random
random.seed(a=17)

rows=random.sample(range(0,len(X.index)-1), int(len(X.index)*0.1))
xtest=imp_X.ix[rows]
ytest=Y.ix[rows]
xtrain=imp_X.drop(rows,axis=0)
ytrain=Y.drop(rows,axis=0)
```

```python
# sort the test data by the house prices (to visualise the results)

testdata=pd.concat([xtest,ytest],axis=1)
testdataSort=testdata.sort_values(['price_doc'])
#testdataSort
```

```python
ytest=testdataSort['price_doc']
xtest=testdataSort.drop(['price_doc'],axis=1)
```

** Create the regression model **

```python
# Fit regression model
params = {'n_estimators': 500, 'max_depth': 8,
          'learning_rate': 0.1,'loss':'ls',
          'min_samples_split': 300, }
clf = ensemble.GradientBoostingRegressor(**params)

clf.fit(xtrain, ytrain)
mse = mean_squared_error(ytest, clf.predict(xtest))
print("MSE: %.4f" % mse)

#  plot the result
plt.figure(figsize=(14,7))
plt.plot(range(ytest.shape[0]),ytest, label='real price')
plt.plot(range(ytest.shape[0]), clf.predict(xtest),label='prediction',alpha=0.4)
plt.legend()

plt.xlabel('Index', fontsize=18)
plt.ylabel('Price (100K RUB)',fontsize=18)
plt.yticks(fontsize=14)
plt.xticks(fontsize=14)
plt.title('House Price Distribution with Prediction',fontsize=18)
plt.show()
```

** Feature importance **

```python
feature_importance = clf.feature_importances_
# make importances relative to max importance
feature_importance = 100.0 * (feature_importance / feature_importance.max())
sorted_idx = np.argsort(feature_importance)
pos = np.arange(sorted_idx.shape[0]) + .5
highfeatures=feature_importance[sorted_idx];
highnames=np.array(xtrain.columns.values)[sorted_idx];
plt.figure(figsize=(14,35))
plt.barh(pos, feature_importance[sorted_idx], align='center')
plt.yticks(pos, np.array(xtrain.columns.values)[sorted_idx])
plt.xlabel('Relative Importance')
plt.title('Variable Importance')
plt.show()
```

** RMSE calculation **

```python
ypredict=clf.predict(xtest)

ydiff=pd.DataFrame(ytest)

ydiff["prediction"]=ypredict
ydiff["differance"]=ydiff["price_doc"]-ydiff["prediction"]

def pow2(x):
    return x**2

ydiff["pow_differance"]=ydiff["differance"].apply(pow2)
```

```python
import math
model_rmse=math.sqrt(sum(ydiff["pow_differance"])*(1/len(ydiff)))
model_rmse
```
