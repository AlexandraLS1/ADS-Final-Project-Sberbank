** Visualizing Datatypes and Missing Vaule Analysis **

```python
# lookin for mising value in each feature in train data.
train.isnull().sum()
```

```python
# count the train type, make a mew data freme
train_DataType=pd.DataFrame(train.dtypes.value_counts()).reset_index().rename(columns={'index':'Vartype',0:'Count'})
train_DataType.head()
```

```python
g=sns.barplot(data=train_DataType,x='Vartype',y="Count")
g.set(xlabel='Variable Type', ylabel='Count',title="Variables Count Across Datatype")
plt.show()
```

** Check wich value is have more NaN**- after we find who is nan we need think what to do with this variales

```python
# get all colmns that have NaN and other 
missing_trainData=train.columns[train.isnull().any()].tolist() # remove all column that dont have NaN, and convert to list of columns name
msno.bar(train[missing_trainData])
```

we get 52 columns that have NaN in com rows,we can see who have more and how less NaN in the data 

**Heat map** 
Other way to see the same.
The correlation heatmap lets as to measire how strongly the presence of one variable positively or negativly affects the presence of other. 

```python
msno.heatmap(train[missing_trainData])
```

Correlation ranges from -1 (if one variable appears the other definitely does not) to 0 (variables appearing or not appearing have no effect on one another) to 1 (if one variable appears the other definitely also does
